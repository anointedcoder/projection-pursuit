```python
# A polynomial function of 3 degrees
def f(x):
    return -2*x**3 + 3*x**2 + 4*x - 5

# Plotting the graph of the function
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 2, 100) # x values from -2 to 2
y = f(x) # y values corresponding to x

plt.plot(x, y, color='blue') # plot the curve
plt.xlabel('x') # label the x-axis
plt.ylabel('f(x)') # label the y-axis
plt.title('A graph of a polynomial function of 3 degrees') # give a title
plt.grid(True) # show the grid
plt.show() # show the graph
```
